#### Graficando los datos del primer pico

```python
# --- PASO 1: Preparar el índice de tiempo en DF1 ---
print("\nProcesando fecha y hora para DF1...")
try:
    # Asegúrate de que las columnas 'Fecha' y 'Hora' existan en DF1
    DF1['Timestamp_str'] = DF1['Fecha'] + ' ' + DF1['Hora']
    DF1['Timestamp'] = pd.to_datetime(DF1['Timestamp_str'], format='%d/%m/%Y %H:%M:%S.%f')
    DF1.set_index('Timestamp', inplace=True)
    print("Índice de tiempo creado para 'DF1'.")
    
    # Eliminamos la columna auxiliar que ya no necesitamos
    DF1.drop(columns=['Timestamp_str'], inplace=True, errors='ignore') 

except KeyError as e:
     print(f"\nError: No se encontró la columna {e} ('Fecha' o 'Hora') en DF1.")
     print(f"Verifica los nombres de columna leídos (pueden ser incorrectos si header=1 fue erróneo).")
     print(f"Columnas actuales en DF1: {DF1.columns.tolist()}")
     exit()
except Exception as e:
    print(f"\nError al procesar fecha/hora en DF1: {e}")
    exit()

# --- PASO 2: Identificar y convertir columnas numéricas en DF1 ---
print("\nIdentificando y convirtiendo columnas numéricas en DF1...")
columnas_a_excluir = ['Fecha', 'Hora'] # Columnas originales que no son datos numéricos directos
columnas_numericas = []
columnas_no_convertidas = []

for col in DF1.columns:
    if col not in columnas_a_excluir:
        try:
            # Intentar convertir a numérico si no lo es ya
            if not pd.api.types.is_numeric_dtype(DF1[col]):
                 DF1[col] = pd.to_numeric(DF1[col], errors='coerce')
            # Si la columna es numérica (originalmente o después de convertirla)
            if pd.api.types.is_numeric_dtype(DF1[col]):
                columnas_numericas.append(col)
            # Si después de intentar convertir sigue sin ser numérica (y no son solo NaNs)
            elif DF1[col].notna().any(): 
                 columnas_no_convertidas.append(col)
        except Exception as e:
            print(f"Advertencia: Error procesando columna '{col}': {e}")
            columnas_no_convertidas.append(col)

if not columnas_numericas:
     print("Error: No se encontraron columnas numéricas válidas en DF1.")
     exit()
if columnas_no_convertidas:
    print(f"Advertencia: Las siguientes columnas no pudieron ser tratadas como numéricas: {columnas_no_convertidas}")

print(f"Columnas numéricas identificadas en DF1: {len(columnas_numericas)}")


# --- PASO 3: Agrupar columnas por prefijo (Mín, Med, Máx) ---
prefijos = set()
sufijos = [' Mín.', ' Med', ' Máx.']
for col in columnas_numericas:
    base = col
    for sufijo in sufijos:
        if col.endswith(sufijo):
            base = col[:-len(sufijo)]
            break
    prefijos.add(base)

print(f"Grupos/Prefijos identificados en DF1: {len(prefijos)}")

# --- PASO 4: Graficar cada grupo de DF1 en una figura separada ---
print("\nGenerando gráficas para cada grupo de DF1...")

for prefijo in sorted(list(prefijos)):
    col_min = f"{prefijo} Mín."
    col_med = f"{prefijo} Med"
    col_max = f"{prefijo} Máx."
    
    cols_a_graficar = {}
    if col_min in columnas_numericas and col_min in DF1.columns: cols_a_graficar['min'] = col_min
    if col_med in columnas_numericas and col_med in DF1.columns: cols_a_graficar['med'] = col_med
    if col_max in columnas_numericas and col_max in DF1.columns: cols_a_graficar['max'] = col_max
    
    if not cols_a_graficar and prefijo in columnas_numericas and prefijo in DF1.columns:
        cols_a_graficar['unica'] = prefijo

    if not cols_a_graficar:
        continue 

    # Crear una NUEVA figura para este grupo
    plt.figure(figsize=(15, 5)) 
    
    # Graficar las columnas encontradas desde DF1
    if 'max' in cols_a_graficar:
        plt.plot(DF1.index, DF1[cols_a_graficar['max']], label='Máx.', color='red', linewidth=1)
    if 'med' in cols_a_graficar:
        plt.plot(DF1.index, DF1[cols_a_graficar['med']], label='Med', color='black', linewidth=1)
    if 'min' in cols_a_graficar:
        plt.plot(DF1.index, DF1[cols_a_graficar['min']], label='Mín.', color='green', linewidth=1)
    if 'unica' in cols_a_graficar:
         plt.plot(DF1.index, DF1[cols_a_graficar['unica']], label=prefijo, color='blue', linewidth=1)

    # Personalizar la gráfica
    plt.title(f'Análisis de: {prefijo} (desde {archivo_1})') # Título indica el origen
    plt.ylabel('Valor') 
    plt.xlabel('Tiempo')
    plt.grid(True)
    if len(cols_a_graficar) > 1 or 'unica' not in cols_a_graficar:
         plt.legend() 
    
    plt.tight_layout()
    plt.show() # Muestra esta figura
```

```python
# --- Constantes ---
I_L1 = 23.0      # Amperios L1
I_L2 = 24.0      # Amperios L2
PF = 0.67        # Factor de Potencia del Sistema Unificado
UMBRAL_CAIDA_V = -0.01 # Umbral para detectar la caída inicial (V)
DURACION_EVENTO_SEGUNDOS = 15 # Duración estimada del evento de arranque (s)
HORAS_DESPUES_PARA_E3 = 12 # Horas después de la caída para calcular E_3 

print(f"Usando Factor de Potencia ÚNICO del Sistema: PF = {PF}")

# --- PASO 0: Cargar y combinar datos ---
print("Cargando y combinando archivos...")
# (Asegúrate que esta parte funcione, REVISANDO header=1)
try:
    archivo_1 = 'Datos_registrados_primer_pico.csv' 
    archivo_2 = 'Datos_registrados_largo.csv'     
    # --- ¡¡REVISA header=1!! --- 
    DF1 = pd.read_csv(archivo_1, delimiter=';', decimal='.', header=1) 
    DF2 = pd.read_csv(archivo_2, delimiter=';', decimal='.', header=1) 
    CONSUMO_ENERGETICO = pd.concat([DF1.copy(), DF2.copy()], ignore_index=True)
    print(f"Archivos cargados y combinados. Total filas: {len(CONSUMO_ENERGETICO)}")
except FileNotFoundError as e: print(f"\nError al cargar archivo: {e}"); exit() 
except Exception as e: print(f"\nError inesperado al cargar/combinar: {e}"); exit()

# --- PASO 1: Preparar índices de tiempo y ordenar ---
print("\nProcesando fecha y hora...")
timestamp_caida = None 
try:
    # Procesar CONSUMO_ENERGETICO
    if 'Fecha' in CONSUMO_ENERGETICO.columns and 'Hora' in CONSUMO_ENERGETICO.columns:
        CONSUMO_ENERGETICO['Timestamp_str'] = CONSUMO_ENERGETICO['Fecha'] + ' ' + CONSUMO_ENERGETICO['Hora']
        CONSUMO_ENERGETICO['Timestamp'] = pd.to_datetime(CONSUMO_ENERGETICO['Timestamp_str'], format='%d/%m/%Y %H:%M:%S.%f')
        CONSUMO_ENERGETICO.set_index('Timestamp', inplace=True)
        CONSUMO_ENERGETICO.drop(columns=['Timestamp_str'], inplace=True, errors='ignore') 
        CONSUMO_ENERGETICO.sort_index(inplace=True)
    else: raise ValueError("CONSUMO_ENERGETICO no tiene 'Fecha'/'Hora'.")

    # Procesar DF1 para encontrar la caída
    if 'Fecha' in DF1.columns and 'Hora' in DF1.columns:
        DF1['Timestamp_str'] = DF1['Fecha'] + ' ' + DF1['Hora']
        DF1['Timestamp'] = pd.to_datetime(DF1['Timestamp_str'], format='%d/%m/%Y %H:%M:%S.%f')
        DF1.set_index('Timestamp', inplace=True)
        DF1.drop(columns=['Timestamp_str'], inplace=True, errors='ignore')
        DF1.sort_index(inplace=True)
    else: raise ValueError("DF1 no tiene 'Fecha'/'Hora'.")
    print("Índices de tiempo creados y ordenados.")

except KeyError as e: print(f"\nError: Columna {e} no encontrada. Revisa header=1."); exit()
except ValueError as e: print(f"\nError en Fecha/Hora: {e}"); exit()
except Exception as e: print(f"\nError al procesar fecha/hora: {e}"); exit()

# --- PASO 2: Asegurar Voltajes y Calcular Potencia/Energía Acumulada ---
print("\nAsegurando voltajes y recalculando Potencia/Energía...")
col_voltaje_ref = 'Tensión L1N Med' 
columnas_voltaje_potencia = [col_voltaje_ref, 'Tensión L2N Med']
try:
    # Verificar y convertir voltajes
    for col in columnas_voltaje_potencia:
        if col not in CONSUMO_ENERGETICO.columns: raise KeyError(f"Columna '{col}' no encontrada.")
        if not pd.api.types.is_numeric_dtype(CONSUMO_ENERGETICO[col]):
            CONSUMO_ENERGETICO[col] = pd.to_numeric(CONSUMO_ENERGETICO[col], errors='coerce')
    CONSUMO_ENERGETICO.dropna(subset=columnas_voltaje_potencia, inplace=True)
    if CONSUMO_ENERGETICO.empty: raise ValueError("No quedan datos válidos tras limpiar voltajes.")

    # Calcular Potencia
    CONSUMO_ENERGETICO['Potencia_L1_kW'] = (CONSUMO_ENERGETICO['Tensión L1N Med'] * I_L1 * PF) / 1000.0 
    CONSUMO_ENERGETICO['Potencia_L2_kW'] = (CONSUMO_ENERGETICO['Tensión L2N Med'] * I_L2 * PF) / 1000.0 
    CONSUMO_ENERGETICO['Potencia_Total_kW'] = CONSUMO_ENERGETICO['Potencia_L1_kW'] + CONSUMO_ENERGETICO['Potencia_L2_kW']

    # Calcular Energía Acumulada
    delta_t_horas = CONSUMO_ENERGETICO.index.to_series().diff().dt.total_seconds() / 3600.0
    potencia_media_intervalo = (CONSUMO_ENERGETICO['Potencia_Total_kW'] + CONSUMO_ENERGETICO['Potencia_Total_kW'].shift(1)) / 2.0
    energia_intervalo = potencia_media_intervalo * delta_t_horas
    CONSUMO_ENERGETICO['Energia_Acumulada_kWh'] = energia_intervalo.fillna(0).cumsum()
    print("Potencia y Energía Acumulada calculadas.")

    # Verificar voltaje en DF1
    if col_voltaje_ref not in DF1.columns: raise KeyError(f"'{col_voltaje_ref}' no encontrada en DF1.")
    if not pd.api.types.is_numeric_dtype(DF1[col_voltaje_ref]):
        DF1[col_voltaje_ref] = pd.to_numeric(DF1[col_voltaje_ref], errors='coerce')
    DF1.dropna(subset=[col_voltaje_ref], inplace=True) 
    if DF1.empty: raise ValueError("DF1 vacío tras limpiar voltaje.")

except KeyError as e: print(f"Error: Columna requerida {e}. Revisa header=1."); exit()
except ValueError as e: print(f"Error: {e}"); exit()
except Exception as e: print(f"Error inesperado en cálculo: {e}"); exit()


# --- PASO 3: Detectar la Primera Caída de Tensión Significativa en DF1 ---
print(f"\nBuscando la primera caída de '{col_voltaje_ref}' > {abs(UMBRAL_CAIDA_V)} V en DF1...")
timestamp_caida = None
try:
    DF1['Voltaje_Diff'] = DF1[col_voltaje_ref].diff()
    caidas_significativas = DF1[DF1['Voltaje_Diff'] < UMBRAL_CAIDA_V]
    
    if caidas_significativas.empty:
        print(f"Advertencia: No se detectó caída > {abs(UMBRAL_CAIDA_V)} V en DF1. Usando primer timestamp de DF1.")
        timestamp_caida = DF1.index[0]
    else:
        timestamp_caida = caidas_significativas.index[0]
        valor_caida = caidas_significativas['Voltaje_Diff'].iloc[0]
        print(f"Primera caída significativa detectada en DF1 en: {timestamp_caida} (Variación: {valor_caida:.2f} V)")

except Exception as e:
    print(f"Error al detectar la caída de tensión: {e}")
    exit()

# --- PASO 4: Calcular Energía DURANTE el Evento de Arranque/Caída (Ahora E_0) ---
print("\nCalculando energía durante el evento de arranque (E_0)...")
E_inicio_evento = None
E_fin_evento = None
E_0 = None # <--- Variable renombrada

if timestamp_caida:
    try:
        ts_inicio_evento = timestamp_caida
        ts_fin_evento = ts_inicio_evento + pd.Timedelta(seconds=DURACION_EVENTO_SEGUNDOS)
        print(f"Periodo del evento definido: {ts_inicio_evento} -> {ts_fin_evento}")

        idx_inicio = CONSUMO_ENERGETICO.index.get_indexer([ts_inicio_evento], method='nearest')[0]
        ts_inicio_real = CONSUMO_ENERGETICO.index[idx_inicio]
        E_inicio_evento = CONSUMO_ENERGETICO.iloc[idx_inicio]['Energia_Acumulada_kWh']

        idx_fin = CONSUMO_ENERGETICO.index.get_indexer([ts_fin_evento], method='nearest')[0]
        ts_fin_real = CONSUMO_ENERGETICO.index[idx_fin]
        E_fin_evento = CONSUMO_ENERGETICO.iloc[idx_fin]['Energia_Acumulada_kWh']
        
        # --- Cálculo de E_0 ---
        E_0 = E_fin_evento - E_inicio_evento # <--- Asignación a E_0

        print(f"Timestamp real inicio evento: {ts_inicio_real} (E_acum={E_inicio_evento:.4f} kWh)")
        print(f"Timestamp real fin evento:   {ts_fin_real} (E_acum={E_fin_evento:.4f} kWh)")
        # --- Imprimiendo E_0 ---
        print(f"Energía consumida DURANTE el evento (E_0): {E_0:.4f} kWh") 

    except IndexError: print("Error: Índice no encontrado para inicio/fin del evento.")
    except Exception as e: print(f"Error al calcular E_0: {e}")


# --- PASO 5: Obtener E_3 (Energía acumulada total ~N horas después de la caída) ---
print(f"\nBuscando Energía Acumulada Total ~{HORAS_DESPUES_PARA_E3} horas después de la caída (E_3)...")
E_3 = None # <--- Variable renombrada y añadida inicialización
if timestamp_caida:
    try:
        # Usa la constante HORAS_DESPUES_PARA_E3
        timestamp_objetivo_horas_despues = timestamp_caida + pd.Timedelta(hours=HORAS_DESPUES_PARA_E3) 
        print(f"Timestamp objetivo (+{HORAS_DESPUES_PARA_E3}h): {timestamp_objetivo_horas_despues}")

        indice_e_horas_despues = CONSUMO_ENERGETICO.index.get_indexer([timestamp_objetivo_horas_despues], method='nearest')[0]
        timestamp_e_horas_despues_real = CONSUMO_ENERGETICO.index[indice_e_horas_despues]
        
        # --- Asignación a E_3 ---
        E_3 = CONSUMO_ENERGETICO.iloc[indice_e_horas_despues]['Energia_Acumulada_kWh'] 

        print(f"Timestamp real más cercano (+{HORAS_DESPUES_PARA_E3}h): {timestamp_e_horas_despues_real}")
        # --- Imprimiendo E_3 ---
        print(f"Valor de Energia Acumulada Total (~{HORAS_DESPUES_PARA_E3}h después) (E_3): {E_3:.4f} kWh") 

    except IndexError: print(f"Error: No se pudo encontrar el índice para +{HORAS_DESPUES_PARA_E3}h.")
    except Exception as e: print(f"Error al obtener E_3: {e}")
else:
    print(f"No se puede calcular E_3 porque no se determinó el timestamp de caída.")

print("\n--- Proceso Finalizado ---")

# Ahora puedes usar las variables E_0 y E_3 más adelante si existen
if E_0 is not None:
    print(f"\nValor final E_0 (Energía durante evento): {E_0:.4f} kWh")
if E_3 is not None:
    print(f"Valor final E_3 (Energía acumulada total {HORAS_DESPUES_PARA_E3}h post-caída): {E_3:.4f} kWh")
```

```python
# --- Constantes ---
I_L1 = 23.0      # Amperios L1
I_L2 = 24.0      # Amperios L2
PF = 0.67        # Factor de Potencia del Sistema Unificado
UMBRAL_CAIDA_V = -0.01 # Umbral para detectar la caída inicial (V)
DURACION_EVENTO_SEGUNDOS = 15 # Duración estimada del evento de arranque (s)
HORAS_DESPUES_PARA_E3 = 12 # Horas después de la caída para calcular E_3
HORAS_DESPUES_ESPECIFICAS = 4.8 # Horas después de la caída para el cálculo específico

print(f"Usando Factor de Potencia ÚNICO del Sistema: PF = {PF}")

# --- PASO 0: Cargar y combinar datos ---
print("Cargando y combinando archivos...")
# (Asegúrate que esta parte funcione, REVISANDO header=1)
try:
    # --- Importar pandas ---
    import pandas as pd # Asegúrate de que pandas está importado

    archivo_1 = 'Datos_registrados_primer_pico.csv'
    archivo_2 = 'Datos_registrados_largo.csv'
    # --- ¡¡REVISA header=1!! ---
    DF1 = pd.read_csv(archivo_1, delimiter=';', decimal='.', header=1)
    DF2 = pd.read_csv(archivo_2, delimiter=';', decimal='.', header=1)
    CONSUMO_ENERGETICO = pd.concat([DF1.copy(), DF2.copy()], ignore_index=True)
    print(f"Archivos cargados y combinados. Total filas: {len(CONSUMO_ENERGETICO)}")
except FileNotFoundError as e: print(f"\nError al cargar archivo: {e}"); exit()
except Exception as e: print(f"\nError inesperado al cargar/combinar: {e}"); exit()

# --- PASO 1: Preparar índices de tiempo y ordenar ---
print("\nProcesando fecha y hora...")
timestamp_caida = None
try:
    # Procesar CONSUMO_ENERGETICO
    if 'Fecha' in CONSUMO_ENERGETICO.columns and 'Hora' in CONSUMO_ENERGETICO.columns:
        CONSUMO_ENERGETICO['Timestamp_str'] = CONSUMO_ENERGETICO['Fecha'] + ' ' + CONSUMO_ENERGETICO['Hora']
        CONSUMO_ENERGETICO['Timestamp'] = pd.to_datetime(CONSUMO_ENERGETICO['Timestamp_str'], format='%d/%m/%Y %H:%M:%S.%f', errors='coerce')
        # Eliminar filas donde Timestamp no pudo ser parseado
        CONSUMO_ENERGETICO.dropna(subset=['Timestamp'], inplace=True)
        CONSUMO_ENERGETICO.set_index('Timestamp', inplace=True)
        CONSUMO_ENERGETICO.drop(columns=['Timestamp_str'], inplace=True, errors='ignore')
        CONSUMO_ENERGETICO.sort_index(inplace=True)
    else: raise ValueError("CONSUMO_ENERGETICO no tiene 'Fecha'/'Hora'.")

    # Procesar DF1 para encontrar la caída
    if 'Fecha' in DF1.columns and 'Hora' in DF1.columns:
        DF1['Timestamp_str'] = DF1['Fecha'] + ' ' + DF1['Hora']
        DF1['Timestamp'] = pd.to_datetime(DF1['Timestamp_str'], format='%d/%m/%Y %H:%M:%S.%f', errors='coerce')
        # Eliminar filas donde Timestamp no pudo ser parseado
        DF1.dropna(subset=['Timestamp'], inplace=True)
        DF1.set_index('Timestamp', inplace=True)
        DF1.drop(columns=['Timestamp_str'], inplace=True, errors='ignore')
        DF1.sort_index(inplace=True)
    else: raise ValueError("DF1 no tiene 'Fecha'/'Hora'.")
    print("Índices de tiempo creados y ordenados.")

except KeyError as e: print(f"\nError: Columna {e} no encontrada. Revisa header=1."); exit()
except ValueError as e: print(f"\nError en Fecha/Hora: {e}"); exit()
except Exception as e: print(f"\nError al procesar fecha/hora: {e}"); exit()

# --- PASO 2: Asegurar Voltajes y Calcular Potencia/Energía Acumulada ---
print("\nAsegurando voltajes y recalculando Potencia/Energía...")
col_voltaje_ref = 'Tensión L1N Med'
columnas_voltaje_potencia = [col_voltaje_ref, 'Tensión L2N Med']
try:
    # Verificar y convertir voltajes
    for col in columnas_voltaje_potencia:
        if col not in CONSUMO_ENERGETICO.columns: raise KeyError(f"Columna '{col}' no encontrada.")
        if not pd.api.types.is_numeric_dtype(CONSUMO_ENERGETICO[col]):
            CONSUMO_ENERGETICO[col] = pd.to_numeric(CONSUMO_ENERGETICO[col], errors='coerce')
    CONSUMO_ENERGETICO.dropna(subset=columnas_voltaje_potencia, inplace=True)
    if CONSUMO_ENERGETICO.empty: raise ValueError("No quedan datos válidos tras limpiar voltajes.")

    # Calcular Potencia
    CONSUMO_ENERGETICO['Potencia_L1_kW'] = (CONSUMO_ENERGETICO['Tensión L1N Med'] * I_L1 * PF) / 1000.0
    CONSUMO_ENERGETICO['Potencia_L2_kW'] = (CONSUMO_ENERGETICO['Tensión L2N Med'] * I_L2 * PF) / 1000.0
    CONSUMO_ENERGETICO['Potencia_Total_kW'] = CONSUMO_ENERGETICO['Potencia_L1_kW'] + CONSUMO_ENERGETICO['Potencia_L2_kW']

    # Calcular Energía Acumulada
    # Asegurar que el índice está ordenado antes de calcular diff()
    CONSUMO_ENERGETICO.sort_index(inplace=True)
    delta_t_horas = CONSUMO_ENERGETICO.index.to_series().diff().dt.total_seconds().fillna(0) / 3600.0
    # Usar la potencia del punto actual para el intervalo que termina en ese punto
    # o la media si se prefiere, pero debe ser consistente.
    # Para el método trapezoidal (media):
    potencia_media_intervalo = (CONSUMO_ENERGETICO['Potencia_Total_kW'] + CONSUMO_ENERGETICO['Potencia_Total_kW'].shift(1)) / 2.0
    # La primera potencia media será NaN, podemos manejarla asignando la potencia actual o 0 para el primer intervalo.
    potencia_media_intervalo.iloc[0] = CONSUMO_ENERGETICO['Potencia_Total_kW'].iloc[0] # O alguna otra lógica
    
    energia_intervalo = potencia_media_intervalo * delta_t_horas
    CONSUMO_ENERGETICO['Energia_Acumulada_kWh'] = energia_intervalo.fillna(0).cumsum()
    print("Potencia y Energía Acumulada calculadas.")

    # Verificar voltaje en DF1
    if col_voltaje_ref not in DF1.columns: raise KeyError(f"'{col_voltaje_ref}' no encontrada en DF1.")
    if not pd.api.types.is_numeric_dtype(DF1[col_voltaje_ref]):
        DF1[col_voltaje_ref] = pd.to_numeric(DF1[col_voltaje_ref], errors='coerce')
    DF1.dropna(subset=[col_voltaje_ref], inplace=True)
    if DF1.empty: raise ValueError("DF1 vacío tras limpiar voltaje.")

except KeyError as e: print(f"Error: Columna requerida {e}. Revisa header=1."); exit()
except ValueError as e: print(f"Error: {e}"); exit()
except Exception as e: print(f"Error inesperado en cálculo: {e}"); exit()


# --- PASO 3: Detectar la Primera Caída de Tensión Significativa en DF1 ---
print(f"\nBuscando la primera caída de '{col_voltaje_ref}' > {abs(UMBRAL_CAIDA_V)} V en DF1...")
timestamp_caida = None
try:
    # Asegurar que DF1 está ordenado por tiempo
    DF1.sort_index(inplace=True)
    DF1['Voltaje_Diff'] = DF1[col_voltaje_ref].diff()
    # Filtrar para encontrar la primera caída después del primer registro (diff() produce NaN en el primero)
    caidas_significativas = DF1[(DF1['Voltaje_Diff'] < UMBRAL_CAIDA_V) & (DF1['Voltaje_Diff'].notna())]

    if caidas_significativas.empty:
        print(f"Advertencia: No se detectó caída > {abs(UMBRAL_CAIDA_V)} V en DF1. Usando primer timestamp de DF1 si existe.")
        if not DF1.empty:
            timestamp_caida = DF1.index[0]
            print(f"Usando primer timestamp de DF1 como referencia: {timestamp_caida}")
        else:
            print("DF1 está vacío, no se puede determinar timestamp_caida.")
            # Considera salir o manejar este caso de otra manera si es crítico
    else:
        timestamp_caida = caidas_significativas.index[0]
        valor_caida = caidas_significativas['Voltaje_Diff'].iloc[0]
        print(f"Primera caída significativa detectada en DF1 en: {timestamp_caida} (Variación: {valor_caida:.2f} V)")

except Exception as e:
    print(f"Error al detectar la caída de tensión: {e}")
    # Considera salir si timestamp_caida es esencial para los siguientes pasos
    # exit()

# --- PASO 4: Calcular Energía DURANTE el Evento de Arranque/Caída (Ahora E_0) ---
print("\nCalculando energía durante el evento de arranque (E_0)...")
E_inicio_evento = None
E_fin_evento = None
E_0 = None # <--- Variable renombrada

if timestamp_caida:
    try:
        ts_inicio_evento = timestamp_caida
        ts_fin_evento = ts_inicio_evento + pd.Timedelta(seconds=DURACION_EVENTO_SEGUNDOS)
        print(f"Periodo del evento definido: {ts_inicio_evento} -> {ts_fin_evento}")

        # Asegurarse de que CONSUMO_ENERGETICO tiene datos y el índice está listo
        if CONSUMO_ENERGETICO.empty or not isinstance(CONSUMO_ENERGETICO.index, pd.DatetimeIndex):
            raise ValueError("CONSUMO_ENERGETICO no está correctamente preparado para la búsqueda de timestamps.")

        # Obtener energía acumulada justo ANTES o EN el inicio del evento
        # Usamos 'asof' para obtener el último valor conocido en o antes de ts_inicio_evento
        # Para ser más precisos con el cálculo de la energía *durante* el evento,
        # necesitamos el valor de Energía Acumulada *antes* del inicio del evento
        # y el valor de Energía Acumulada *al final* del evento.

        # Energía acumulada al inicio del periodo del evento
        # Buscamos el punto más cercano ANTES o IGUAL al ts_inicio_evento
        energia_antes_inicio_evento_series = CONSUMO_ENERGETICO['Energia_Acumulada_kWh'].asof(ts_inicio_evento - pd.Timedelta(microseconds=1))
        if pd.isna(energia_antes_inicio_evento_series):
            # Si no hay datos antes, podríamos tomar el primer valor disponible si está cerca
            # O el valor en el timestamp más cercano (podría ser el mismo ts_inicio_evento)
            idx_inicio_aprox = CONSUMO_ENERGETICO.index.get_indexer([ts_inicio_evento], method='nearest')[0]
            E_inicio_evento = CONSUMO_ENERGETICO.iloc[idx_inicio_aprox]['Energia_Acumulada_kWh']
            ts_inicio_real = CONSUMO_ENERGETICO.index[idx_inicio_aprox]
            print(f"Advertencia: No se encontró un punto exacto antes de {ts_inicio_evento}. Usando el más cercano: {ts_inicio_real} para E_inicio_evento.")
        else:
            E_inicio_evento = energia_antes_inicio_evento_series
            ts_inicio_real = CONSUMO_ENERGETICO.index[CONSUMO_ENERGETICO.index <= ts_inicio_evento][-1] # Aproximación del timestamp real
        
        # Energía acumulada al final del periodo del evento
        # Buscamos el punto más cercano EN o DESPUÉS de ts_fin_evento
        idx_fin = CONSUMO_ENERGETICO.index.get_indexer([ts_fin_evento], method='nearest')[0]
        ts_fin_real = CONSUMO_ENERGETICO.index[idx_fin]
        E_fin_evento = CONSUMO_ENERGETICO.iloc[idx_fin]['Energia_Acumulada_kWh']

        # --- Cálculo de E_0 ---
        if E_inicio_evento is not None and E_fin_evento is not None:
            E_0 = E_fin_evento - E_inicio_evento
            print(f"Timestamp real considerado para inicio evento (o anterior más cercano): {ts_inicio_real} (E_acum_antes={E_inicio_evento:.4f} kWh)")
            print(f"Timestamp real más cercano fin evento: {ts_fin_real} (E_acum_en_fin={E_fin_evento:.4f} kWh)")
            print(f"Energía consumida DURANTE el evento (E_0): {E_0:.4f} kWh")
        else:
            print("No se pudo calcular E_0 porque E_inicio_evento o E_fin_evento es None.")

    except IndexError: print("Error: Índice no encontrado para inicio/fin del evento en CONSUMO_ENERGETICO.")
    except ValueError as e: print(f"Error de valor al buscar timestamps para E_0: {e}")
    except Exception as e: print(f"Error al calcular E_0: {e}")
else:
    print("No se puede calcular E_0 porque no se determinó el timestamp de caída.")


# --- PASO 5: Obtener E_3 (Energía acumulada total ~N horas después de la caída) ---
print(f"\nBuscando Energía Acumulada Total ~{HORAS_DESPUES_PARA_E3} horas después de la caída (E_3)...")
E_3 = None
consumo_desde_caida_hasta_E3 = None

if timestamp_caida:
    try:
        timestamp_objetivo_horas_despues = timestamp_caida + pd.Timedelta(hours=HORAS_DESPUES_PARA_E3)
        print(f"Timestamp objetivo (+{HORAS_DESPUES_PARA_E3}h): {timestamp_objetivo_horas_despues}")

        if CONSUMO_ENERGETICO.empty or not isinstance(CONSUMO_ENERGETICO.index, pd.DatetimeIndex):
            raise ValueError("CONSUMO_ENERGETICO no está correctamente preparado.")

        # Energía acumulada en el timestamp de la caída (o el más cercano disponible)
        idx_caida = CONSUMO_ENERGETICO.index.get_indexer([timestamp_caida], method='nearest')[0]
        E_en_caida = CONSUMO_ENERGETICO.iloc[idx_caida]['Energia_Acumulada_kWh']
        ts_caida_real = CONSUMO_ENERGETICO.index[idx_caida]

        # Energía acumulada en el timestamp objetivo
        indice_e_horas_despues = CONSUMO_ENERGETICO.index.get_indexer([timestamp_objetivo_horas_despues], method='nearest')[0]
        timestamp_e_horas_despues_real = CONSUMO_ENERGETICO.index[indice_e_horas_despues]
        E_3_acumulada_total = CONSUMO_ENERGETICO.iloc[indice_e_horas_despues]['Energia_Acumulada_kWh']
        E_3 = E_3_acumulada_total # E_3 es la energía *acumulada total* en ese punto

        consumo_desde_caida_hasta_E3 = E_3_acumulada_total - E_en_caida

        print(f"Timestamp real de caída usado para E_3: {ts_caida_real} (E_acum_en_caida={E_en_caida:.4f} kWh)")
        print(f"Timestamp real más cercano (+{HORAS_DESPUES_PARA_E3}h): {timestamp_e_horas_despues_real}")
        print(f"Valor de Energia Acumulada Total (~{HORAS_DESPUES_PARA_E3}h después) (E_3): {E_3:.4f} kWh")
        print(f"Consumo desde la caída hasta ~{HORAS_DESPUES_PARA_E3}h después: {consumo_desde_caida_hasta_E3:.4f} kWh")

    except IndexError: print(f"Error: No se pudo encontrar el índice para +{HORAS_DESPUES_PARA_E3}h.")
    except ValueError as e: print(f"Error de valor al buscar timestamps para E_3: {e}")
    except Exception as e: print(f"Error al obtener E_3: {e}")
else:
    print(f"No se puede calcular E_3 porque no se determinó el timestamp de caída.")

# --- PASO 6: Obtener Energía después de HORAS_DESPUES_ESPECIFICAS (ej. 4.8 horas) ---
print(f"\nBuscando Energía Acumulada y Consumo después de {HORAS_DESPUES_ESPECIFICAS} horas desde la caída...")
E_especifico_acumulada = None
consumo_hasta_E_especifico = None

if timestamp_caida:
    try:
        timestamp_objetivo_especifico = timestamp_caida + pd.Timedelta(hours=HORAS_DESPUES_ESPECIFICAS)
        print(f"Timestamp objetivo (+{HORAS_DESPUES_ESPECIFICAS}h): {timestamp_objetivo_especifico}")

        if CONSUMO_ENERGETICO.empty or not isinstance(CONSUMO_ENERGETICO.index, pd.DatetimeIndex):
            raise ValueError("CONSUMO_ENERGETICO no está correctamente preparado.")

        # Energía acumulada en el timestamp de la caída (o el más cercano disponible)
        # Podríamos reutilizar E_en_caida y ts_caida_real si ya se calcularon y son válidos
        # Por claridad, lo recalculamos o verificamos
        idx_caida_esp = CONSUMO_ENERGETICO.index.get_indexer([timestamp_caida], method='nearest')[0]
        E_en_caida_esp = CONSUMO_ENERGETICO.iloc[idx_caida_esp]['Energia_Acumulada_kWh']
        ts_caida_real_esp = CONSUMO_ENERGETICO.index[idx_caida_esp]

        # Energía acumulada en el timestamp objetivo específico
        indice_e_especifico = CONSUMO_ENERGETICO.index.get_indexer([timestamp_objetivo_especifico], method='nearest')[0]
        timestamp_e_especifico_real = CONSUMO_ENERGETICO.index[indice_e_especifico]
        E_especifico_acumulada = CONSUMO_ENERGETICO.iloc[indice_e_especifico]['Energia_Acumulada_kWh']

        consumo_hasta_E_especifico = E_especifico_acumulada - E_en_caida_esp

        print(f"Timestamp real de caída usado para cálculo específico: {ts_caida_real_esp} (E_acum_en_caida={E_en_caida_esp:.4f} kWh)")
        print(f"Timestamp real más cercano (+{HORAS_DESPUES_ESPECIFICAS}h): {timestamp_e_especifico_real}")
        print(f"Energía Acumulada Total (~{HORAS_DESPUES_ESPECIFICAS}h después): {E_especifico_acumulada:.4f} kWh")
        print(f"Consumo desde la caída hasta ~{HORAS_DESPUES_ESPECIFICAS}h después: {consumo_hasta_E_especifico:.4f} kWh")

    except IndexError: print(f"Error: No se pudo encontrar el índice para +{HORAS_DESPUES_ESPECIFICAS}h.")
    except ValueError as e: print(f"Error de valor al buscar timestamps para el cálculo específico: {e}")
    except Exception as e: print(f"Error al obtener Energía después de {HORAS_DESPUES_ESPECIFICAS}h: {e}")
else:
    print(f"No se puede calcular la energía después de {HORAS_DESPUES_ESPECIFICAS}h porque no se determinó el timestamp de caída.")


print("\n--- Proceso Finalizado ---")

# Ahora puedes usar las variables E_0, E_3, consumo_desde_caida_hasta_E3, E_especifico_acumulada y consumo_hasta_E_especifico
if E_0 is not None:
    print(f"\nValor final E_0 (Energía durante evento): {E_0:.4f} kWh")
if E_3 is not None: # E_3 es la energía acumulada total
    print(f"Valor final E_3 (Energía acumulada total {HORAS_DESPUES_PARA_E3}h post-caída): {E_3:.4f} kWh")
if consumo_desde_caida_hasta_E3 is not None:
    print(f"Consumo neto desde caída hasta ~{HORAS_DESPUES_PARA_E3}h: {consumo_desde_caida_hasta_E3:.4f} kWh")
if E_especifico_acumulada is not None:
    print(f"Energía acumulada total {HORAS_DESPUES_ESPECIFICAS}h post-caída: {E_especifico_acumulada:.4f} kWh")
if consumo_hasta_E_especifico is not None:
    print(f"Consumo neto desde caída hasta ~{HORAS_DESPUES_ESPECIFICAS}h: {consumo_hasta_E_especifico:.4f} kWh")
```
